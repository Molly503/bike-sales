```python
import pandas as pd
import matplotlib.pyplot as pltb
sales = pd.read_csv('/Users/jieliang/bikes_sales_project/web_sales.csv')
```

```python
sales
```

```python
sales['product_subcategory'].unique()
```

```python
sales['product_subcategory'].isin(['Road Bikes','Mountain Bikes','Touring Bikes'])
```

```python
bike_sales=sales[sales['product_subcategory'].isin(['Road Bikes','Mountain Bikes','Touring Bikes'])]
```

```python
bike_sales
```

```python
bike_sales.groupby('product_subcategory')['SalesAmount'].sum().reset_index()
```

```python
accessory_sales = sales[~sales['product_subcategory'].isin(['Road Bikes','Mountain Bikes','Touring Bikes'])]
```

```python
accessory_sales
```

```python
accessory_sales.groupby('product_subcategory')['SalesAmount'].sum().reset_index()
```

```python
accessory_sales_grp=accessory_sales.groupby('product_subcategory')\
['SalesAmount'].sum().reset_index().sort_values(by='SalesAmount')
```

```python
accessory_sales_grp
```

```python
bike_sales_grp=bike_sales.groupby('product_subcategory')\
['SalesAmount'].sum().reset_index().sort_values(by='SalesAmount', ascending=False)
```

```python
bike_sales_grp
```

```python
plt.pie(bike_sales_grp['SalesAmount'],
        labels=bike_sales_grp['product_subcategory'])
plt.show()
```

```python
plt.barh(accessory_sales_grp['product_subcategory'],accessory_sales_grp['SalesAmount']
        )
plt.show()
```

```python
sales['OrderDate']=pd.to_datetime(sales['OrderDateKey'],format='%Y%m%d')
sales['MonthYear']=pd.to_datetime(sales['OrderDate'].dt.strftime('%Y-%m'))
```

```python
sales
```

```python
monthly_sales = sales.groupby('MonthYear')['SalesAmount'].sum().reset_index()
```

```python
monthly_sales
```

```python
plt.plot(monthly_sales['MonthYear'],monthly_sales['SalesAmount'])
plt.show()
```

```python
country_sales=sales.groupby('SalesTerritoryCountry')['SalesAmount']\
.sum().reset_index().sort_values(by='SalesAmount', ascending = False)
```

```python
plt.bar(country_sales['SalesTerritoryCountry'],country_sales['SalesAmount'])
```

```python
plt.style.use( 'ggplot')
```

```python
plt.figure(figsize=(16, 12))

plt.subplot(2,2,1)
plt.plot(monthly_sales['MonthYear'], monthly_sales['SalesAmount'], marker='.', linestyle='--', linewidth=2)
plt.xticks(rotation=45, fontsize=8, color='black')
plt.title('Sales by Month')
plt.xlabel('Month-Year', fontsize=10)
plt.ylabel('Sales', fontsize=10)

plt.subplot(2,2,2)
plt.barh(accessory_sales_grp['product_subcategory'], accessory_sales_grp['SalesAmount'])
plt.title('Accessory Sales')

plt.subplot(2,2,3)
plt.pie(bike_sales_grp['SalesAmount'], labels=bike_sales_grp['product_subcategory'], autopct='%1.1f%%')
plt.title('Bike Sales')

plt.subplot(2,2,4)
plt.bar(country_sales['SalesTerritoryCountry'], country_sales['SalesAmount'])
plt.xticks(rotation=45, fontsize=8, color='black')
plt.title('Country Sales')
plt.xlabel('Countries', fontsize=10)
plt.ylabel('Sales', fontsize=10)

plt.subplots_adjust(wspace=0.5, hspace=0.5)  # Adjust subplots spacing
plt.tight_layout(pad=3.0)  # Apply tight layout with padding
plt.show()
```
